```python
import numpy as np
from scipy import stats
from scipy.stats import norm, t
import matplotlib.pyplot as plt
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
# Definindo os parmetros da distribuio normal
media = 15  # Mdia em kg
desvio_padrao = 0.1  # Desvio padro em kg

# Definindo o limite superior
limite_superior = 15.2  # kg

# Calculando a probabilidade usando a funo de distribuio acumulada (CDF)
probabilidade = 1 - norm.cdf(limite_superior, loc=media, scale=desvio_padrao)

print("A probabilidade de um pacote ter mais de 15.2 kg  de aproximadamente {:.2%}".format(probabilidade))
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Definindo os parmetros da distribuio normal
media = 15  # Mdia em kg
desvio_padrao = 0.1  # Desvio padro em kg

# Obtendo o percentil de ordem 95%
percentil_95 = norm.ppf(0.95, loc=media, scale=desvio_padrao)

print("O percentil de ordem 95% para o peso dos pacotes  de aproximadamente {:.2f} kg.".format(percentil_95))
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
# Valores contidos no objeto Y
Y = np.linspace(-6, 6, 200)

# Distribuio normal padro
prob_Y1 = norm.cdf(Y, loc=0, scale=1)

# Distribuio t-Student com 5 graus de liberdade
prob_Y2 = t.cdf(Y, df=5)

# Imprimir as probabilidades
print("Probabilidades para Y1 (Normal Padro):")
for valor, probabilidade in zip(Y, prob_Y1):
    print("Prob(Y1 < {:.2f}) = {:.4f}".format(valor, probabilidade))

print("\nProbabilidades para Y2 (t-Student com 5 graus de liberdade):")
for valor, probabilidade in zip(Y, prob_Y2):
    print("Prob(Y2 < {:.2f}) = {:.4f}".format(valor, probabilidade))

```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
# Grfico para Y1 (Normal Padro)
plt.figure(figsize=(10, 6))
plt.plot(Y, prob_Y1, label='Y1 (Normal Padro)')
plt.title('Probabilidade de Y1 ser menor que cada valor em Y')
plt.xlabel('Valores de Y')
plt.ylabel('Probabilidade')
plt.legend()
plt.grid(True)
plt.show()

# Grfico para Y2 (t-Student com 5 graus de liberdade)
plt.figure(figsize=(10, 6))
plt.plot(Y, prob_Y2, label='Y2 (t-Student com 5 graus de liberdade)', color='orange')
plt.title('Probabilidade de Y2 ser menor que cada valor em Y')
plt.xlabel('Valores de Y')
plt.ylabel('Probabilidade')
plt.legend()
plt.grid(True)
plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
# Valores contidos no objeto P
P = np.linspace(0, 1, 20)

# Distribuio normal padro
percentil_Y1 = norm.ppf(P, loc=0, scale=1)

# Distribuio t-Student com 5 graus de liberdade
percentil_Y2 = t.ppf(P, df=5)

# Imprimir os percentis
print("Percentis para Y1 (Normal Padro):")
for valor_P, percentil in zip(P, percentil_Y1):
    print("Percentil Y1 para P={:.2f}: {:.4f}".format(valor_P, percentil))

print("\nPercentis para Y2 (t-Student com 5 graus de liberdade):")
for valor_P, percentil in zip(P, percentil_Y2):
    print("Percentil Y2 para P={:.2f}: {:.4f}".format(valor_P, percentil))

```

6. Faa um grfico com os dados do item anterior.

```python
# Grfico para Y1 (Normal Padro)
plt.figure(figsize=(10, 6))
plt.plot(P, percentil_Y1, label='Y1 (Normal Padro)')
plt.title('Percentis para Y1 em relao aos valores em P')
plt.xlabel('Valores de P')
plt.ylabel('Percentil')
plt.legend()
plt.grid(True)
plt.show()

# Grfico para Y2 (t-Student com 5 graus de liberdade)
plt.figure(figsize=(10, 6))
plt.plot(P, percentil_Y2, label='Y2 (t-Student com 5 graus de liberdade)', color='orange')
plt.title('Percentis para Y2 em relao aos valores em P')
plt.xlabel('Valores de P')
plt.ylabel('Percentil')
plt.legend()
plt.grid(True)
plt.show()
```
