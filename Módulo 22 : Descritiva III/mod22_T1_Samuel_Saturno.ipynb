#### 1. Baixando dados de ativos
    1. Baixe os dados de três ativos da sua escolha do Yahoo Finance, correspondente a um período de pelo menos 2 anos até a data mais recente.
    2. Salve esses dados para que você possa utilizá-los na próxima tarefa.

```python
##time

# Lista dos ativos criptomoedas de interesse
ativos = ['BTC-USD', 'ETH-USD', 'USDT-USD']

# Obtendo dados do Yahoo Finance para os ativos especificados a partir de 1º de janeiro de 2022
yf = pdr.get_data_yahoo(tickers=ativos, start='2022-01-01')
yf
```

```python
yf['Adj Close'].isna().sum()
```

```python
# Exportando os dados para um arquivo CSV chamado 'ativos_cripto_top3.csv'
yf.to_csv(path_or_buf='ativos_cripto_top3.csv')

# Lendo os dados do arquivo CSV e armazenando em um DataFrame
df = pd.read_csv(filepath_or_buffer='ativos_cripto_top3.csv',
                 header=[0,1],  # Usando a primeira e segunda linha como cabeçalho
                 index_col=0)   # Usando a primeira coluna como índice
df  # Exibindo o DataFrame
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
# Empilhando o DataFrame para converter as colunas em um índice hierárquico
df = df.stack()

# Exibindo o DataFrame resultante após a operação de empilhamento
df
```

```python
# Calculando a faixa de preço e adicionando-a como uma nova coluna chamada 'Price Range'
df['Price Range'] = df['High'] - df['Low']

# Desempilhando o DataFrame para reverter a operação anterior
df = df.unstack()

# Extraindo a coluna 'Price Range' do DataFrame resultante
price_range = df['Price Range']

price_range  # Exibindo a série contendo a faixa de preço
```

```python
# Empilhando a coluna 'Price Range' para transformá-la em um DataFrame
df_range = df['Price Range'].stack().reset_index()

# Renomeando as colunas do DataFrame resultante
df_range.columns = ['data', 'ativo', 'valor']

# Exibindo o DataFrame resultante após as operações
df_range
```

```python
# Criando um gráfico de linha com os dados do DataFrame df_range
# Configurando a data como o eixo x, o valor como o eixo y e o ativo como a cor das linhas
# Utilizando o template 'plotly_white' para o estilo do gráfico
fig = px.line(data_frame=df_range,
              x='data',
              y='valor',
              color='ativo',
              template='plotly_white')

# Exibindo o gráfico
fig.show()
```

### Insight:

Os valores exibidos representam a variação de preços dos ativos 'BTC-USD', 'ETH-USD' e 'USDT-USD'. A amplitude das flutuações dessas
criptomoedas é visualizada no gráfico acima. Podemos notar que o ativo 'BTC-USD' apresenta a maior amplitude, seguido por 'ETH-USD' e, por último, 'USDT-USD'. Esses dados de volatilidade são referentes ao período desde 2022 até o momento atual.
