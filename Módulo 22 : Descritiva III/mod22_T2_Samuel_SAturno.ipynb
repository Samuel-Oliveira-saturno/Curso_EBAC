### Importando as Bibliotecas

````python
import pandas as pd

import plotly.express as px

from plotly import graph_objects
from plotly.graph_objs import Layout
````

#### 1. Carregue os dados da tarefa anterior

````python
# Carregar o arquivo CSV contendo os dados dos três principais ativos de criptomoedas
# O argumento 'filepath_or_buffer' especifica o caminho do arquivo
# 'header=[0,1]' indica que a primeira e a segunda linhas do arquivo CSV serão usadas como cabeçalhos de coluna
# 'index_col=0' define a primeira coluna do arquivo CSV como o índice do DataFrame
df = pd.read_csv(filepath_or_buffer='/content/ativos_cripto_top3.csv',
                 header=[0,1],
                 index_col=0)

# Imprimir o DataFrame carregado
df
````

#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
# Calcula a média móvel de 90 períodos dos preços de fechamento ajustados (Adj Close)
# 'rolling(window=90)' especifica a janela de 90 períodos para calcular a média móvel
df_ma = df['Adj Close'].rolling(window=90).mean()

# Retorna a série com a média móvel de 90 períodos
df_ma
````

````python
# Calcula o desvio padrão móvel de 90 períodos dos preços de fechamento ajustados (Adj Close)
# 'rolling(window=90)' especifica a janela de 90 períodos para calcular o desvio padrão móvel
df_std = df['Adj Close'].rolling(window=90).std()

# Retorna a série com o desvio padrão móvel de 90 períodos
df_std
````

````python
# Transforma o DataFrame da média móvel em um formato adequado para plotagem
# 'stack()' transforma as colunas do DataFrame em índices, criando uma série de dados
# 'reset_index()' recria o DataFrame com os índices redefinidos
df_ma_adjclose = df_ma.stack().reset_index()

# Renomeia as colunas para 'data', 'ativo' e 'valor'
df_ma_adjclose.columns = ['data', 'ativo', 'valor']

# Cria um gráfico de linhas interativo usando o Plotly
# 'data_frame' define o DataFrame a ser usado
# 'x' especifica os dados para o eixo x (data)
# 'y' especifica os dados para o eixo y (valor)
# 'color' define a variável para colorir as linhas (ativo)
# 'template' define o esquema de cores do gráfico
fig = px.line(data_frame=df_ma_adjclose,
              x='data',
              y='valor',
              color='ativo',
              template='plotly_white')

# Mostra o gráfico
fig.show()
````

### Insight 01:

O gráfico exibe o histórico de preços dos ativos BTC, ETH e USDT de julho de 2022 a janeiro de 2024. Durante esse intervalo, observamos flutuações significativas no preço do Bitcoin, caracterizando sua volatilidade intrínseca. Podemos verificar os valores de forma mais precisa ao apontar o cursor sobre a trajetória da linha de cada ativo.

De modo geral, podemos inferir que o preço do Bitcoin iniciou acima de  60.000 em julho de 2022, caindo para abaixo de  10.000 em algum momento anterior a janeiro de 2023. Posteriormente, houve uma recuperação, elevando o preço para mais de  60.000 em janeiro de 2024.

É crucial lembrar que o Bitcoin é conhecido por sua alta volatilidade, resultando em rápidas variações de preço.

````python
# Transforma o DataFrame do desvio padrão em um formato adequado para plotagem
# 'stack()' transforma as colunas do DataFrame em índices, criando uma série de dados
# 'reset_index()' recria o DataFrame com os índices redefinidos
df_std_adjclose = df_std.stack().reset_index()

# Renomeia as colunas para 'data', 'ativo' e 'valor'
df_std_adjclose.columns = ['data', 'ativo', 'valor']

# Cria um gráfico de linhas interativo usando o Plotly
# 'data_frame' define o DataFrame a ser usado
# 'x' especifica os dados para o eixo x (data)
# 'y' especifica os dados para o eixo y (valor)
# 'color' define a variável para colorir as linhas (ativo)
# 'template' define o esquema de cores do gráfico
fig = px.line(data_frame=df_std_adjclose,
              x='data',
              y='valor',
              color='ativo',
              template='plotly_white')

# Mostra o gráfico
fig.show()
````

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
# Seleciona as últimas 60 linhas do DataFrame para o ativo 'BTC-USD'
# 'xs(key='BTC-USD', level=1, axis=1)' retorna as colunas (nível 1) correspondentes ao ativo 'BTC-USD'
# 'tail(n=60)' retorna as últimas 60 linhas do DataFrame selecionado
df_last60 = df.xs(key='BTC-USD',
                  level=1,
                  axis=1).tail(n=60)

# Imprime a quantidade de linhas do período selecionado
print(f'Quantidade de linhas do período selecionado: {df_last60.shape[0]}.\n')

# Retorna as últimas 60 linhas do DataFrame para o ativo 'BTC-USD'
df_last60
````

````python
# Define o layout do gráfico
# 'paper_bgcolor' define a cor de fundo do papel como transparente
# 'plot_bgcolor' define a cor de fundo do gráfico como branco
# 'height' define a altura do gráfico como 600 pixels
layout = Layout(paper_bgcolor='rgba(0,0,0,0)',
                plot_bgcolor='white',
                height=600)

# Define os dados do gráfico em formato de candlestick para o ativo 'BTC-USD'
graph = {
    'x': df_last60.index,
    'open': df_last60.Open,
    'close': df_last60.Close,
    'high': df_last60.High,
    'low': df_last60.Low,
    'type': 'candlestick',
    'name': 'BTC-USD',
    'showlegend': True
}

# Cria o gráfico usando a biblioteca plotly.graph_objects
grafico = graph_objects.Figure(data=[graph],
                               layout_title='Índice Bitcoin',
                               layout=layout)

# Atualiza os eixos x e y do gráfico para exibir as grades de fundo
grafico.update_xaxes(showgrid=True,
                     gridwidth=1,
                     gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True,
                     gridwidth=1,
                     gridcolor='LightGrey')

# Retorna o gráfico
grafico
````
