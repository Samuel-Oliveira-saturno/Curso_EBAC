# Importando as Bibliotecas

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Criando o Dataframe

```python
# Gerando os dados
np.random.seed(123)

p1 = np.random.random(50) *10
p1[p1>10] = 10
p1[p1<0] = 0

p2 = p1 + np.random.normal(0,1,50)
p2[p2>10] = 10
p2[p2<0] = 0

df = pd.DataFrame({'p1': p1, 'p2': p2})
df.head()
```



A célula abaixo gera dados de notas de duas provas, p1 e p2, conforme em aula, mas com mais alunos.

## 1. Reduzindo correlações



<h3><font  color="brown"><strong> A)  Obtenha a média e a diferença através de uma multiplicação matricial </strong></font></h3>

```python
# Calculando média e evolução
df['média'] = df.mean(axis=1)
df['diferença'] = df.p2-df.p1
df['soma'] = df.p2+df.p1
df.style.format(precision=2, decimal=',')
```

<h3><font  color="brown"><strong> B) Calcule a correlação entre os dados da p1 e p2 </strong></font></h3>

```python
# Calculando a correlação entre p1 e p2
correlation = df['p1'].corr(df['p2'])

# Imprimindo o resultado
print("Correlação entre p1 e p2:", correlation)

```

<h3><font  color="brown"><strong> C) Calcule a correlação entre os dados transformados (média e diferença) </strong></font></h3>


```python
# Criando as variáveis média e diferença
df['media'] = (df['p1'] + df['p2']) / 2
df['diferenca'] = df['p1'] - df['p2']

# Calculando a correlação entre as variáveis média e diferença
correlacao = df[['media', 'diferenca']].corr().iloc[0, 1]

print("Correlação entre média e diferença:", correlacao)
```

<h3><font  color="brown"><strong> D) Comente as correlações observadas </strong></font></h3>

<h3><font  color=""><i> No item B podemos verificar que a correlação entre p1 e p2 é bastante alta, aproximadamente 0.896. Isso indica uma forte relação linear positiva entre essas duas variáveis. Em outras palavras, quando o valor de p1 aumenta, espera-se que o valor de p2 também aumente, e vice-versa. Essa correlação sugere que as duas variáveis estão relacionadas de alguma forma, o que pode ser útil para entender o comportamento dos dados e para realizar análises posteriores.
</i></font></h3>

<h3><font  color=""><i>
No item C temos que a correlação entre os dados transformados (média e diferença), que é aproximadamente -0.133, indica uma relação linear fraca negativa entre essas duas variáveis.
</i></font></h3>

<h3><font  color=""><i>
Especificamente:
</i></font></h3>


<h3><font  color=""><i>
* A correlação próxima de zero indica que não há uma relação linear forte entre a média e a diferença.
</i></font></h3>

<h3><font  color=""><i>
* A correlação negativa sugere que, em geral, quando a média aumenta, a diferença tende a diminuir um pouco, e vice-versa. No entanto, essa relação é fraca, o que significa que não é muito confiável para fazer previsões sobre a diferença com base na média, e vice-versa.
</i></font></h3>

<h3><font  color=""><i>
Portanto, embora haja uma relação, é importante considerar que ela não é tão forte quanto a correlação entre p1 e p2, o que indica que a transformação dos dados pode ter alterado a natureza da relação entre as variáveis. Isso ressalta a importância de compreender a natureza dos dados transformados e como essas transformações afetam suas propriedades estatísticas.
</i></font></h3>

## 2. Soma e diferença

<h3><font  color="brown"><strong>
Realize agora uma transformação semelhante, mas que ao invés da média, dê a soma da p1 e p2. Construa o gráfico e conclua que diferenças visuais você observa.
</strong></font></h3>


```python
# Criando uma nova coluna com a soma de p1 e p2
df['p1_plus_p2'] = df['p1'] + df['p2']

# Criando o gráfico de dispersão
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='p1', y='p2', hue='p1_plus_p2', palette='viridis', legend='brief', s=80)

# Adicionando uma linha de referência
plt.plot([0, 10], [0, 10], '--', color='red', linewidth=2)

# Adicionando rótulos e título
plt.xlabel('p1', fontsize=12)
plt.ylabel('p2', fontsize=12)
plt.title('Soma de p1 e p2', fontsize=14)
plt.legend(title='p1 + p2')

# Mostrando o gráfico
plt.grid(True)
plt.tight_layout()
plt.show()


```

<h3><font  color=""><strong>
Insight:
</strong></font></h3>

<h3><font  color=""><i>
Optamos pelo gráfico de dispersão para explicar a relação entre p1 e p2, com a cores distintas representando a soma de p1 e p2, podemos observar alguns pontos importantes:
</i></font></h3>

<h3><font  color=""><strong>
Padrão de dispersão:
</strong></font></h3>


<h3><font  color=""><i>
Os pontos no gráfico estão distribuídos em torno de uma linha diagonal, o que indica uma relação positiva entre p1 e p2. Isso confirma a alta correlação positiva entre p1 e p2 que calculamos anteriormente.
</i></font></h3>


<h3><font  color=""><strong>
Coloração dos pontos:
</strong></font></h3>


<h3><font  color=""><i>
A cor dos pontos representa a soma de p1 e p2. Podemos observar que os pontos mais escuros (indicando somas maiores) estão concentrados nas áreas onde os valores de p1 e p2 são altos. Isso faz sentido, já que a soma de duas variáveis maiores resultará em um valor de soma maior.
</i></font></h3>

<h3><font  color=""><strong>
Linha de referência:
</strong></font></h3>



<h3><font  color=""><i>
A linha vermelha pontilhada representa a linha onde p1 é igual a p2. Podemos ver que a maioria dos pontos está acima dessa linha, o que indica que, em geral, p1 tende a ser maior que p2.
</i></font></h3>

<h3><font  color=""><i>
Com base nessas observações, podemos concluir que há uma forte relação positiva entre p1 e p2, e que a soma dessas variáveis tende a ser maior em áreas onde p1 e p2 são maiores individualmente.
