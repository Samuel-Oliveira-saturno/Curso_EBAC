````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
````

### Carregando a base de dados "previsao_de_renda.csv'

````python
# importando e visualizando os dados do DataFrame
df = pd.read_csv('previsao_de_renda.csv')
df.head()
````

````python
# Limpando os dados do DataFrame
df.drop(columns=['Unnamed: 0'], inplace=True)
````

````python
# Selecionar apenas as variáveis quantitativas do DataFrame
df_quantitative = df.select_dtypes(include=['int64', 'float64'])
````

1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
# Dividir o DataFrame em grupos com base na posse de imóvel
grupo_posse_imovel = df.groupby('posse_de_imovel')['renda'].apply(list)

# Visualizando os dados
grupo_posse_imovel
````

````python
# Teste t para comparar as médias de renda entre os grupos
t_stat_imovel, p_valor_imovel = stats.ttest_ind(grupo_posse_imovel[0], grupo_posse_imovel[1])

# Visualizando os dados de teste
print("Posse de imóvel:")
print("   Teste t-valor:", t_stat_imovel)
print("   Valor p:", p_valor_imovel)
````

````python
# Dividir o DataFrame em grupos com base na posse de veículo
grupo_posse_veiculo = df.groupby('posse_de_veiculo')['renda'].apply(list)

# visualizando os grupo
grupo_posse_veiculo
````

````python
# Teste t para comparar as médias de renda entre os grupos
t_stat_veiculo, p_valor_veiculo = stats.ttest_ind(grupo_posse_veiculo[0], grupo_posse_veiculo[1])

# Visualizando o valores de teste
print("Posse de veículo:")
print("   Teste t-valor:", t_stat_veiculo)
print("   Valor p:", p_valor_veiculo)
````

## Insights

Com base nos resultados dos testes t para as variáveis qualitativas "posse de imóvel" e "posse de veículo", podemos extrair os seguintes insights:

### Posse de Imóvel:

* O teste t-valor é -3.65.
*O valor p é 0.00026.
*O valor p muito baixo (menor que 0.05) sugere que há uma diferença estatisticamente significativa nas médias de renda entre os grupos definidos pela posse de imóvel.
*Isso indica que a posse de imóvel pode ser um fator discriminante importante na determinação da renda dos indivíduos no conjunto de dados. Pessoas com posse de imóvel tendem a ter uma renda significativamente diferente daquelas sem posse de imóvel.


### Posse de Veículo:

* O teste t-valor é -7.78.
* O valor p é 7.62e-15.
* O valor p extremamente baixo (praticamente zero) indica uma diferença estatisticamente significativa nas médias de renda entre os grupos definidos pela posse de veículo.
* Isso sugere que a posse de veículo é um preditor muito forte da renda no conjunto de dados. Indivíduos com posse de veículo têm uma renda

### consideravelmente diferente daqueles sem posse de veículo.


Esses insights sugerem que tanto a posse de imóvel quanto a posse de veículo são variáveis importantes na predição da renda das pessoas no conjunto de dados. Essas descobertas podem ter implicações significativas em análises mais detalhadas ou na construção de modelos preditivos de renda.
