# 1. Qual a principal diferença entre um *behaviour score* e um *credit score*?

### O credit score avalia o risco de inadimplência com base no histórico geral de crédito em várias instituições. O behaviour score analisa o comportamento financeiro atual de um cliente com produtos de uma instituição específica. Enquanto o primeiro é estático e histórico, o segundo é dinâmico e focado no presente.

# 2. Por que é tão comum se estabelecer janelas de desempenho de 12 meses, e 12 safras para o desenvolvimento de modelos?


### Janelas de 12 meses capturam sazonalidades completas, como variações de comportamento em períodos específicos do ano. Elas fornecem uma amostra estável e suficiente para análises robustas. Além disso, são um padrão de mercado amplamente aceito.

# 3. Qual a diferença entre amostra *out of time* e amostra *out of sample*?

### A amostra out of time separa dados de períodos distintos, testando o modelo em cenários futuros. A amostra out of sample divide os dados aleatoriamente, sem considerar a temporalidade. Ambas validam o modelo, mas com enfoques diferentes.

# 4. Se os dados da variavel resposta estão sistematicamente corrompidos para duas das safras de desenvolvimento, o que você faria?

### Excluiria ou corrigiria as safras corrompidas, dependendo da gravidade do problema e disponibilidade de dados confiáveis. Caso excluídas, ajustaria o modelo para compensar a perda de representatividade. Se corrigidas, aplicaria técnicas de imputação ou validação cruzada para garantir a qualidade.
