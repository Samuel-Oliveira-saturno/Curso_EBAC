````python
#%%time

# Carrega o arquivo CSV utilizando a biblioteca pandas
y_train = (pd.read_csv(filepath_or_buffer=_y_train, #  Especifica o caminho do arquivo CSV
                             header=None) # Indica que o CSV não possui uma linha de cabeçalho
                   .squeeze('columns')) # o Método 'squeeze' transforma o Dataframe em uma série

# Exibi a série 
y_train
````

````python
%%time  

# Carrega um arquivo CSV para os IDs de sujeitos dos dados de teste (subject_test) usando a biblioteca pandas
# O arquivo é lido a partir do caminho especificado pela variável _subject_test
# Não há uma linha de cabeçalho no arquivo, então o parâmetro header=None é usado
# A função squeeze é usada para transformar o DataFrame resultante em uma série
subject_test = (pd.read_csv(filepath_or_buffer=_subject_test, 
                            header=None)
                   .squeeze('columns'))

# Carrega um arquivo CSV para os dados de teste (X_test) usando a biblioteca pandas
# O arquivo é lido a partir do caminho especificado pela variável _X_test
# Não há uma linha de cabeçalho no arquivo, então o parâmetro header=None é usado
# O parâmetro names=features indica que o DataFrame resultante deve ter colunas nomeadas conforme os valores em 'features'
# O parâmetro delim_whitespace=True é usado para indicar que os valores no arquivo são separados por espaços em branco
# A função assign é usada para adicionar a coluna 'subject_test' ao DataFrame
X_test = (pd.read_csv(filepath_or_buffer=_X_test, 
                      header=None, 
                      names=features, 
                      delim_whitespace=True)
             .assign(subject_test=subject_test))

# Define a coluna 'subject_test' como o índice do DataFrame X_test
# O parâmetro drop=True indica que a coluna utilizada como índice deve ser removida do DataFrame
# O parâmetro append=True indica que o índice 'subject_test' deve ser adicionado como nível adicional ao índice existente
# O parâmetro inplace=True indica que a operação deve ser realizada diretamente no DataFrame X_test
X_test.set_index(keys='subject_test', 
                 drop=True, 
                 append=True, 
                 inplace=True)

# Exibe o DataFrame X_test
X_test
````

````python
#%%time

# Carrega o arquivo CSV utilizando a biblioteca pandas
y_test = (pd.read_csv(filepath_or_buffer=_y_test, #  Especifica o caminho do arquivo CSV
                             header=None) # Indica que o CSV não possui uma linha de cabeçalho
                   .squeeze('columns')) # o Método 'squeeze' transforma o Dataframe em uma série

# Exibi a série 
y_test
````

### 2. Divisão da base em Treino, Validação e Teste

A base já se encontra dividida em Treino e Validação. O que vamos fazer então é extrair uma base de Validação da base de Treino.

Extraia 25% da base de treino como base de validação.

````python
# Divide os dados de treinamento X_train e y_train em conjuntos de treinamento e validação
# O parâmetro test_size=0.25 define que 25% dos dados serão utilizados para validação, enquanto 75% serão para treinamento
# O parâmetro random_state=42 garante a reproducibilidade dos resultados, fixando a semente aleatória
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.25, random_state=42)

# Imprime as dimensões dos conjuntos de treinamento e validação
print('X_train:', X_train.shape)
print('y_train:', y_train.shape)
print('X_valid:', X_valid.shape)
print('y_valid:', y_valid.shape)

# Como os conjuntos de teste X_test e y_test já foram carregados previamente, apenas imprime suas dimensões
print('X_test:', X_test.shape)
print('y_test:', y_test.shape)
````

### 3. Melhores 3 variáveis

Rode uma árvore com profundidade máxima igual a 4 para prever a atividade humana com todas as variáveis.
Observe a importância das variáveis e considere as 3 variáveis com maior importância para os próximos passos.
Dica: utilize o atributo ```clf.feature_importances_``` da árvore treinada.

````python
# Cria uma instância do classificador DecisionTreeClassifier com max_depth=4 e random_state=42
# Treina o classificador utilizando os dados de treinamento X_train e y_train
clf = DecisionTreeClassifier(max_depth=4, random_state=42).fit(X=X_train, y=y_train)

# Cria um DataFrame com as variáveis e suas importâncias de acordo com o classificador treinado
# Ordena o DataFrame pelas importâncias das variáveis em ordem decrescente
three_best_vars = (pd.DataFrame(data={'var': X_train.columns, 
                                      'feature_importances_': clf.feature_importances_})
                     .sort_values(by='feature_importances_', ascending=False))

# Imprime as variáveis e suas importâncias, ordenadas pela importância
print(three_best_vars)

# Seleciona as três melhores variáveis com base nas importâncias
cols = three_best_vars['var'].values[:3]

# Imprime as três melhores variáveis selecionadas
print('3 melhores variáveis:', cols)
````

### 4. Construa uma árvore com as 3 melhores variáveis

Utilizando as três variáveis encontradas acima, construa uma árvore de decisão. Encontre o melhor ```ccp_alpha``` utilizando a base de validação, conforme a estrutura que vimos em aula.

````python
%%time

# Calcula o caminho de poda por complexidade de custo para a árvore de decisão utilizando os dados de treinamento reduzidos às três melhores variáveis
path = DecisionTreeClassifier(random_state=42).cost_complexity_pruning_path(X=X_train[cols], y=y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

# Inicializa uma lista para armazenar os classificadores para cada valor de ccp_alpha
clfs = []

# Itera sobre os valores de ccp_alphas
for ccp_alpha in ccp_alphas:
    # Cria uma instância do classificador DecisionTreeClassifier com o valor atual de ccp_alpha
    clf = DecisionTreeClassifier(random_state=42, ccp_alpha=ccp_alpha).fit(X=X_train[cols], y=y_train)
    
    # Adiciona o classificador treinado à lista de classificadores clfs
    clfs.append(clf)

# Calcula os scores de treinamento para cada classificador na lista clfs
train_scores = [clf.score(X=X_train[cols], y=y_train) for clf in clfs]

# Calcula os scores de validação para cada classificador na lista clfs
valid_scores = [clf.score(X=X_valid[cols], y=y_valid) for clf in clfs]

# Cria um gráfico para visualizar a acurácia em função do parâmetro de complexidade ccp_alpha
fig, ax = plt.subplots()
ax.set_xlabel(xlabel='Alpha')
ax.set_ylabel(ylabel='Acurácia')
ax.set_title(label='Acurácia X Alpha do conjunto de dados de treino e validação')
ax.plot(ccp_alphas, train_scores, marker='o', label='Treino', drawstyle='steps-post')
ax.plot(ccp_alphas, valid_scores, marker='o', label='Validação', drawstyle='steps-post')
ax.legend()
plt.show()

````

````python
# Encontra o índice da melhor árvore na base de validação com base nos scores de validação
# O índice é encontrado utilizando a função index para achar o índice do primeiro valor máximo encontrado na lista valid_scores
# Como valid_scores está ordenado do menor para o maior ccp_alpha, a melhor árvore está no índice encontrado
best_tree_index = len(valid_scores) - valid_scores[::-1].index(max(valid_scores)) - 1

# Acessa a melhor árvore de decisão da lista de classificadores clfs utilizando o índice encontrado
best_tree = clfs[best_tree_index]

# Imprime a acurácia da melhor árvore na base de validação, formatada para exibir duas casas decimais
print(f'Acurácia da melhor árvore na base de validação: {valid_scores[best_tree_index] * 100:.2f}')

# Retorna a melhor árvore de decisão para uso posterior
best_tree
````

### 5. Avaliação do modelo

Avalie a árvore encontrada no item anterior na base de testes.

````python
# Calcula a acurácia da melhor árvore na base de testes
accuracy_test = best_tree.score(X=X_test[cols], 
                                y=y_test)


# Imprime a acurácia da melhor árvore na base de testes, formatada para exibir duas casas decimais
print(f'Acurácia da melhor árvore na base de testes: {accuracy_test * 100:.2f}')


# Cria e exibe a matriz de confusão para visualizar o desempenho do classificador na base de testes
ConfusionMatrixDisplay.from_estimator(estimator=best_tree, 
                                      X=X_test[cols], 
                                      y=y_test);
````
